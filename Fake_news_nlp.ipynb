```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn import svm
```

```python
fake=pd.read_csv(r"C:\Users\Dhrub\Downloads\archive (19)\Fake.csv")
real=pd.read_csv(r"C:\Users\Dhrub\Downloads\archive (19)\True.csv")
```

```python
fake.head()
```

```python
real.head()
```

```python
fake["label"] = 0
real["label"] = 1

df = pd.concat([fake, real])
```

```python
df.head()
```

```python
df['subject'].value_counts()
```

```python
df["text"] = df["text"].str.lower()
import re
import string
#remove punctuation
df['text'] = df['text'].str.translate(str.maketrans('', '', string.punctuation))
#remove numbers
df['text'] = df['text'].str.replace(r'\d+', '', regex=True)
# Remove HTML tags
df['text'] = df['text'].str.replace(r'<.*?>', '', regex=True)
# Remove URLs
df['text'] = df['text'].str.replace(r'http\S+|www\S+', '', regex=True)
#Remove emojis
df['text'] = df['text'].str.replace(r'[\U00010000-\U0010ffff]','', regex=True)    
```

```python
df.head()
```

```python
st=set(stopwords.words('english'))
```

```python
df['text'] = df['text'].apply(lambda x: " ".join([w for w in x.split() if w not in st]))
```

```python
df['text']
```

```python
X=df['text']
y=df['label']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
tf=TfidfVectorizer()
X_train_tf=tf.fit_transform(X_train)
X_test_tf=tf.transform(X_test)
```

```python
lb=LogisticRegression()
lb.fit(X_train_tf,y_train)
y_p3=lb.predict(X_test_tf)
accuracy_score(y_p3,y_test)
```

```python
print(confusion_matrix(y_test, y_p3))
```

```python
print(classification_report(y_test, y_p3))
```

```python
user_news = input("Enter news article: ")

user_vector = tf.transform([user_news])

prediction = lb.predict(user_vector)
probability = lb.predict_proba(user_vector)

if prediction[0] == 0:
    print("Fake News")
    print("Confidence:", round(probability[0][0] * 100, 2), "%")
else:
    print("Real News")
    print("Confidence:", round(probability[0][1] * 100, 2), "%")
```

```python
import joblib

joblib.dump(lb, "fake_news_model.pkl")
joblib.dump(tf, "tfidf_vectorizer.pkl")
```
